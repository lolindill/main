# load data as runInfo



```python
import requests
def get_raw_urls(repo_owner, repo_name, branch, directory):
    base_url = f"https://api.github.com/repos/{repo_owner}/{repo_name}/contents/{directory}?ref={branch}"
    raw_base_url = f"https://raw.githubusercontent.com/{repo_owner}/{repo_name}/{branch}"

    response = requests.get(base_url)
    if response.status_code == 200:
        files = response.json()
        raw_urls = [f"{raw_base_url}/{file['path']}" for file in files if file['type'] == 'file']
        return raw_urls
    else:
        print(f"Failed to fetch directory contents: {response.status_code}")
        return []
repo_owner = 'KU81-ComSci-ProjectEnd-GameCard'
repo_name = 'main'
branch = 'master'
directory = 'Predictor/data/raw_data'
raw_urls = get_raw_urls(repo_owner, repo_name, branch, directory)
raw_urls = [string.replace('#', '%23') for string in raw_urls]
```

```python
json_data = list()
for file in raw_urls:
  with urllib.request.urlopen(file) as url:
    json_data.extend(json.loads(url.read().decode()))
```

```python
import pandas as pd
df = pd.DataFrame(json_data)
```

```python
runInfo = df['event'].apply(pd.Series)
```

```python
def upgrade_lst(lst):
  new_lst = list()
  for item in lst:
    if(item == 'Searing Blow'):
      new_lst.append('Searing Blow')
    else:
      new_lst.append(item+'+1')
  return new_lst
```

```python
ALL_CARD = IRONCLAD_CARDS + COLORLESS_CARDS + CURSE_CARDS + upgrade_lst(IRONCLAD_CARDS) + upgrade_lst(COLORLESS_CARDS)
```

# Filtering runInfo

speciel case for 'Searing Blow' that can upgrade many time

```python
import re
def clear_SearingBlowPlus(lst):
    return [re.sub(r'Searing Blow\+\d+', 'Searing Blow', item) for item in lst]
runInfo['master_deck'] = runInfo['master_deck'].apply(clear_SearingBlowPlus)
```

filter only ironclad run

```python
def is_subset(words, lst):
    return all(word in lst for word in words)
ic_run = runInfo[runInfo['master_deck'].apply(lambda x: is_subset(x, ALL_CARD))]
```
